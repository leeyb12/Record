# 파이썬 기초

## Print 기초

```python
# 문자
print("주피터랩이 처음이에요.")
# 숫자
print(1234)
```

```python
# 문자 + 문자
print("주피터랩이 " + "처음이에요.")
# 숫자 + 숫자
print(1234 + 2345)
```

```python
# 문자 + 문자(숫자x)
print("주피터랩이 " + "1234")
```

- **문자+숫자에러**



```python
# 문자 X 숫자 = 반복
print("안녕" * 5)
```

<hr/>

```python
# print 심화1
print("=" * 20, end='') # 마지막에 end=''를 이용하여 엔터를 삭제함.
print("안녕" * 5, end='')
print("=" * 20, end='')
```

```python
# print 심화2
print("안녕", "하시", "겠습니까", "?", sep='-')
print(1, 2, 3, 4, 5, sep='/')
```

```python
# print 응용1
import time as t
print("잠온다")
t.sleep(1.5)
print("배가 불러서일까?")
t.sleep(1.5)
print("배가 고파서일까?")
```

```python
# print 응용2
import time as t
print("기대하세요.", end='\r')
t.sleep(1.5)
print("드디어~~~", end='\r')
t.sleep(1.5)
print("개방합니다!", end='\r')
```

## 숫자형

```python
# 정수형
a = 123
b = -321
c = 0
```

```python
print(a)
print(b)
print(c)
```

```python
# 실수형
aa = 3.141592
bb = -12.34
cc = 3.14E10
dd = 3.14e-10
ee = 2**10
print(aa)
print(bb)
print(cc)
print(dd)
print(ee)
```

```python
# 2진수(bin)
aaa = 0b1101
# 8진수(oct)
bbb = 0o177
# 16진수(hex)
ccc = 0x8ff
print(aaa)
print(bbb)
print(ccc)
```

```python
ss = 5
dd = 7
ss ** dd
print(ss ** dd)
```

## 사칙연산

```python
# 사칙연산 예
sa = 77
san = 13
print('덧셈:', sa + san)
print('뺄셈:', sa - san)
print('곱셈:', sa * san)
print('나눗셈:', sa / san)
print('몫:', sa // san)
print('나머지:', sa % san)
print('승:', sa ** san)
```

```python
print(10*18**2+2*11)
```

```python
print(14 // 3)
print(14 % 3)
```

```python
# 복합 연산자
a = 1 # 초기값
a = a + 1
print(a)
```

```python
a += 1
print(a)
```

```python
b = 10
```

```python
b -= 1
print(b)
```
