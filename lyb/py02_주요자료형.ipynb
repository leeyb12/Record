### 리스트 관련함수

```python
# append - 추가하기
a = []
print(a)
a.append(1)
print(a)
a.append(2)
print(a)
a.append([11, 22, 33])
print(a)
a.append(4)
print(a)
```

```python
# sort - 리스트 정렬
a = [5, 43, 5, 6, 56, 3, 2, 4, 1]
b = ['gf', 'f', 'u', 'a', 'e', 'f', 'z', 'c']
a.sort()
print(a)
print(b)
```

```python
# reverse - 리스트 뒤집기
a = [5, 43, 5, 6, 56, 3, 2, 4, 1]
b = ['gf', 'f', 'u', 'a', 'e', 'f', 'z', 'c']
print(a[::-1])
print(b[::-1])
a.reverse()
b.reverse()
print(a)
print(b)
```

```python
# sort와 reverse 응용
a = [5, 43, 5, 6, 56, 3, 2, 4, 1]
a.sort()
a.reverse()
print(a)
b = [5, 43, 5, 6, 56, 3, 2, 4, 1]
b.sort(reverse=True) # True는 내림차순, False 오름차순이 기본값
print(b)
```

```python
# sorted - 원본 유지하기
aa = [66, 7, 8, 99, 0]
print(sorted(aa))
print(sorted(aa, reverse=True))
print(aa)
```

```python
# sorted 응용하기
fruit = ['banana', 'kiwi', 'orange', 'melon', 'apple', 'strawberry']
print(sorted(fruit))
print(sorted(fruit, key=len)) # 글자 길이의 오름차순 
print(sorted(fruit, key=len, reverse=True)) # 글자 길이의 내림차순
```

```python
# insert 리스트에 요소삽입 *
a = [1, 2, 3, 4, 5, 6]
a.insert(0, 11)
print(a)
a.insert(3, 22)
print(a)
a.insert(4, [1, 2, 3])
print(a)
a.insert(len(a), 777)
print(a)
```

```python
# pop 리스트 요소 꺼내기 **
a = [1, 2, 3, 4, 5]
out = a.pop(0) # 값을 넣으면 인덱스의 요소를 반환하고 그 요소는 삭제한다.
print(out, a) # 값이 없을 때(기본값)는 맨 뒤에서 부터 꺼내면서 삭제.
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
```

```python
# count 리스트에 포함된 요소 개수세기
a = [2, 3, 4, 6] * 2 + [1, 2, 3] * 3
print(a, a.count(4))
```

```python
# extend 리스트 확장
a = [1, 2, 3]
a.extend([4, 5])
print(a)
```

### 응용문제

```python
# 장바구니 초기화
bag = [] # bag = list()
```

```python
# append를 이용한 자료넣기
item = input("필요한 항목을 넣어주세요.")
bag.append(item)
print("장바구니에는 {0} 들이 들어있습니다.".format(bag))
```

```python
# FIFO
hand = bag.pop(0)
print("%s를 꺼내서 장바구니에 %s 가 남았습니다." %(hand, bag))
```

```python
# 선택 출력

> "바구니에 ['호박', '감자', '소주', '맥주', '피자', '치킨']이 있습니다."
> "꺼내고 싶은 물품을 입력하세요"
> "바구니에서 ??? 을 꺼내어 [ ~, ~, ~, ~]가 남았습니다."
```

```python
basket = ['호박', '감자', '소주', '맥주', '피자', '치킨']

print(f"바구니에 {basket}이 있습니다.")
bring = input("꺼내고 싶은 물품을 입력하세요: ") 

item = basket.pop(basket.index(bring))
print(f"바구니에서 {item}을 꺼내어 {list(basket)}가 남았습니다.")
```
